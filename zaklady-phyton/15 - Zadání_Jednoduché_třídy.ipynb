````python
class Clovek:
    def __init__(self, jmeno:str, prijmeni, telefon, email, adresa, pojistovna):
        self.jmeno = jmeno
        self.prijmeni = prijmeni
        self.telefon = telefon
        self.email = email
        self.adresa = adresa
        self.pojistovna = pojistovna

    def get_jmeno(self):
        return self.jmeno
    def set_jmeno(self, nove_jmeno):
        self.jmeno = nove_jmeno
        print("jmeno bylo zmeneno")

    def get_prijmeni(self):
        return self.prijmeni
    def set_prijmeni(self,prijmeni):
        self.prijmeni = prijmeni
        print("prijmeni bylo zmeneno")

    def get_telefon(self):
        return self.telefon
    def set_telefon(self, telefon):
        self.tefon = telefon
        print("telefon byl zmenen")

    def get_email(self):
        return self.email
    def set_email(self, email):
        self.email = email
        print("email byl zmenen")

    def get_adresa(self):
        return self.adresa
    def set_adresa(self, adresa):
        self.adresa = adresa
        print("adresa byla zmenena")

    def get_pojistovna(self):
        return self.pojistovna
    def set_pojistovna(self, pojistovna):
        self.pojistovna = pojistovna
        print("pojistovna byla zmenena")

    def __str__(self):
        return f"""
        jmeno: {self.jmeno} 
        prijmeni: {self.prijmeni} 
        telefon: {self.telefon}
        email: {self.email} 
        adresa: {self.adresa} 
        pojistovna: {self.pojistovna}"""

clovek1 = Clovek("Daria", "Bobnova", 123456, "dariabob@gmail.com", "FFG", "VZP")
clovek2 = Clovek("Karel", "Vidovy", 58945, "karlv@gmail.com", "hFG", "VZDR")

print(clovek1)
clovek1.set_prijmeni("cococ")
print(clovek1)
print(clovek2.get_jmeno()) 


        
````

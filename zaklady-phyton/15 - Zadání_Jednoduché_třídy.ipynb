Vytvořte třídu Pocasi, která bude mít atributy:

1.   vítr:float

Vytvořte metody, které budou moc získávat a měnit tyto atributy. Také vytvořte metodu, která vám vypíše vše tomto počasí


````python
from enum import Enum
class Pocasi:
    def __init__(self, vitr, podminky, teplota, vlhkost):
        self.vitr = vitr
        self.podminky = podminky
        self.teplota = teplota
        self.vlhkost = vlhkost

class Ppodminky(Enum):
    SLUNECNO = "slunečno"
    ZATAZENO="zataženo"
    BOURKA="bouřka"
    SNIH ="sníh"

    def get_vitr(self):
        return self.vitr
    def set_vitr(self, nvitr):
        self.vitr = nvitr

    def get_podminky(self):
        return self.podminky
    def set_podminky(self, npodminky):
        self.podminky = npodminky

    def get_teplota(self):
        return self.teplota
    def set_teplota(self, nteplota):
        self.teplota = nteplota

    def get_vlhkost(self):
        return self.vlhkost
    def set_vlhkost(self, nvlhkost):
        self.vlhkost = nvlhkost

    def __str__(self):
        return f""" 
        vitr: {self.vitr} km/h
        podminky: {self.podminky}
        teplota: {self.teplota} C
        vlhkost: {self.vlhkost} %"""
    #TODO
    p1 = Pocasi(31,Ppodminky.SLUNECNO.value, 25, 14 )
    print(p1)
````

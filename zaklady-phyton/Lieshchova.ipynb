* Do x uložte 3.
* Do Y uložte 7.

* Poté umocněte toto číslo na 2
* Tento výsledek odmocněte druhou mocninou
* Výsdelek vynásobte třema a přičtěte 1

Vypište výsledek a určete, zda je číslo sudé nebo liché.


````python
#TODO
x = 3
y = 7
print("Pro x =", x)
print("Pro y =", y)
z= x+y
q= z**2
b= q**0.5
a= b*3+1
if a%2 == 0:
    print("Výsledek je", a, "sudé" )
else:
    print("Výsledek je", a, "liché" )



````

Napište do programu hlášku: "Vyberte matematickou operaci: 1 pro sčítání a 2 pro odčítání". Pokud uživatel zadá 1, bude zadávat dvě čísla, která se poté sečtou. Pokud 2 tak to samé ale pro rozdíl.
Výsledek vypište ve formátu:

"první +/- druhý = výsledek"


````python
#TODO
print("Vyberte matematickou operaci: 1 pro sčítání a 2 pro odčítání")
x = int(input("operace 1 = +, 2 = -"))
if x == 1:
    a= int(input("cislo 1"))
    b= int(input("cislo 1"))
    print("Zadejte první číslo:", a)
    print("Zadejte druhé číslo:", b)
    y = a+b
    print(a,"+", b, "=", y)
elif x == 2:
    a= int(input("cislo 1"))
    b= int(input("cislo 1"))
    print("Zadejte první číslo:", a)
    print("Zadejte druhé číslo:", b)
    y = a-b
    print(a,"-", b, "=", y)
else:
    print("vyberte 1 nebo 2")
````

# 3

Vytvoř program, který požádá uživatele o zadání počtu bodů (0 - 100).

Na základě počtu bodů program přiřadí známku podle těchto pravidel:
* 90 až 100 bodů: "Výborně (1)"
* 75 až 89 bodů: "Chvalitebně (2)"
* 60 až 74 bodů: "Dobře (3)"
* 50 až 59 bodů: "Dostatečně (4)"
* 0 až 49 bodů: "Nedostatečně (5)"

Pokud uživatel zadá číslo mimo rozsah 0-100, program vypíše chybovou hlášku: "Neplatný počet bodů."

````python
#TODO
x = int(input("zsdej pocet bodu"))
print("Zadej počet bodů (0-100):", x)
if x > 100:
    print("Neplatný počet bodů")
elif x < 0:
    print("Neplatný počet bodů")
elif x > 90:
    print("Známka: Výborně (1)")
elif x > 75 < 89 :
    print("Známka: Chvalitebně (2)")
elif x > 60 < 74 :
    print("Známka: Dobře (3)")
elif x > 50 < 59 :
    print("Známka: Dostatečně (4)")
elif x > 0 < 49 :
    print("Známka: Nedostatečně (5)")
````

Vytvořte aplikaci, která bude přijímat tři vstupy uživatele.

* prvním bude výška v cm
* druhý váha v kg
* třetí věk

Vypočítejte mi BMI index pro tyto údaje. BMI = (váha v kg) děleno (výška v metrech na druhou)

Výpis napište ve formátu:

Uživatel ve věku "věk" má BMI: "BMI" což znamená Nadváha/ Podváha atd.

viz hodnoty níže.
* 0 až 18,5: Hodnota podváhy
* 18,5 až 25: Hodnota ideální váhy
* 25 až 30: Hodnota nadváhy
* 30 až 40: Hodnota obezity prvního stupně
* 40 a více: Hodnota těžké obezity

````python
vyska = float(input("vyska"))
print(vyska)
vaha = float(input("vaha v kg"))
print(vaha)
vek = float(input("vek"))
print(vek)
bmi = vaha/vyska**2
print("bmi = ", bmi)
if bmi < 18.5:
    print("Hodnota podváhy")
elif bmi > 18.5 < 25:
    print("Hodnota ideální váhy")
elif bmi > 25 < 30:
    print("Hodnota nadváhy")
elif bmi > 30 < 40:
    print("Hodnota obezity prvního stupně")
elif bmi > 40:
    print("Hodnota těžké obezity")
````

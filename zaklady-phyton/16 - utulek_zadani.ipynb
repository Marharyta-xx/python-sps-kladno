# 2
Vytvořte třídu Zvire, která bude obsahovat tyto vlasnosti
* Jméno
* Hmotnost

a metody budou pro get a set a ToString a metodu zarvi()

```python
class Zvire:
  # Každé zvíře má své jméno a hmotnost
  def __init__(self, jmeno, hmotnost):
    if hmotnost < 0:
      raise ValueError("Hmotnost Zvířete nemůže být záporná")
    self.jmeno = jmeno
    self.hmotnost = hmotnost

  @property
  def jmeno(self):
    return self.__jmeno
  @jmeno.setter
  def jmeno(self, jmeno):
    self.__jmeno = jmeno.upper()

  @property
  def hmotnost(self):
    return self.__hmotnost
  @hmotnost.setter
  def hmotnost(self, hmotnost):
    self.__hmotnost = max(0, hmotnost)

  # Každé zvíře umí zařvat
  def zarvi(self):
    print("Huááá hlasitostí")

  def __str__(self):
    return "Volej na mě {} a vážím {} kg".format(self.jmeno, self.hmotnost)

z1 = Zvire("Anička", 30)
print(z1)
z1.zarvi()
```
